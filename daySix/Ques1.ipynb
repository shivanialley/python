```python
""" Ques1   Create a numpy array x containing data representing time(in hrs) spend on studing : [1,2,3,4,5]
            Create a  numpy array y containing corresponding data representing study score(out of 100) :
            [50,60,70,80,90]
            Plot a line graph with blue solid line
            Overlay (on top of it) a scatter plot with red markers"""

import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([50,60,70,80,90])
plt.title("Time on study")
plt.xlabel("Time(in hrs)")
plt.ylabel("Study Score(out of 100)")
plt.plot(xpoints,ypoints,color='b')
plt.scatter(xpoints,ypoints,color='r')
plt.show()
```
