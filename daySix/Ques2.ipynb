```python
""" Ques2 Create a numpy array categories representing different subject : 
["math","science","history","english","arts"]
              Create a numpy array scores with average score(out of 100) for each subject :[85,70,75,90,80]
              Plot a bar graph representing subject & its score
              Colour each bar differently
              Add label, title & modify grid line to display a dash style"""
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["Math","Science","History","English","Arts"])
y=np.array([85,70,75,90,80])
mycolor=["red","yellow","green","pink","black"]
plt.title("Average Score")
plt.xlabel("Subjects")
plt.ylabel("Score(out of 100)")
plt.bar(x,y,width=0.25,color=mycolor)
plt.grid()
plt.show()
```
