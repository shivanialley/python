```python
#5. Which TV show was in the top 10 for a long time?
import pandas as pd

df=pd.read_csv("Netflix_top_10.csv")
print("No. of row & column",df.shape,"\n")
print("First Eight rows\n",df.head(8),"\n")
print("Arithmetic Operation\n",df.describe(),"\n")
print("Show ""Friend""in the top10\n")
print(df[(df["weekly_rank"]<10)&(df["show_title"]=="Friends")],"\n")
print("TV show was in the top 10 for a long time\n")
print(df[(df["cumulative_weeks_in_top_10"]<10)&(df["category"]=="TV (English)")])
```

```python
#7. Which movie was at number 1 position for the month of July?
import pandas as pd

df=pd.read_csv("Netflix_top_10.csv")
print("No. of row & column",df.shape,"\n")
print("First Eight rows\n",df.head(8),"\n")
print("Arithmetic Operation\n",df.describe(),"\n")
print("Show ""Friend""in the top10\n")
print(df[(df["weekly_rank"]<10)&(df["show_title"]=="Friends")])
print("TV show was in the top 10 for a long time\n")
print(df[(df["cumulative_weeks_in_top_10"]<10)&(df["category"]=="TV (English)")])
print("Movies was in the top 10 for a long time\n")
print(df[(df["cumulative_weeks_in_top_10"]<10)&(df["category"]=="Films (English)")])
df['month'] = pd.DatetimeIndex(df['week']).month
print("Movie was at number 1 position for the month of july\n")
print(df[(df["month"] == 7)&(df["cumulative_weeks_in_top_10"]==1)&(df["category"]=="Films (English)")])
```
