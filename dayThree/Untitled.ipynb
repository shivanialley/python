```python
import datetime

def add_employee():
    empid = int(input("Enter the employee id: "))
    emp_name = input("Enter the employee name: ")
    designation = input("Enter the employee designation: ")

    # Handling the input of Joining date format
    while True:
        try:
            joining_date = input("Enter the employee Joining date (YYYY-MM-DD): ")
            datetime.datetime.strptime(joining_date, "%Y-%m-%d")
            break
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

    salary = float(input("Enter the salary: "))

    employee = {}
    employee["id"] = empid
    employee["name"] = emp_name
    employee["job_role"] = designation
    employee["join_date"] = joining_date
    employee["sal"] = salary
    
    employees[empid] = employee

employees = {}

num = 5

# Add initial 5 employees
for i in range(num):
    add_employee()

print("Employee Details:")
for empid, emp_details in employees.items():
    print(f"Employee ID: {empid}")
    for key, value in emp_details.items():
        print(f"{key.capitalize()}: {value}")
    print()

# Find the senior most employee based on joining date
def get_joining_date(emp_id):
    return datetime.datetime.strptime(employees[emp_id]["join_date"], "%Y-%m-%d")

senior_employee_id = max(employees, key=get_joining_date)
senior_employee_details = employees[senior_employee_id]

print("\nSenior Most Employee:")
print(f"Employee ID: {senior_employee_id}")
for key, value in senior_employee_details.items():
    print(f"{key.capitalize()}: {value}")
```

```python
def validate_prefix():
    while True:
        prefix = input("Enter the prefix (Mr., Mrs., Ms.): ").strip().title()
        if prefix in ["Mr.", "Mrs.", "Ms."]:
            return prefix
        else:
            print("Invalid prefix. Please enter one of 'Mr.', 'Mrs.', or 'Ms.'.")

def generate_full_name(prefix, first_name, last_name):
    return f"{prefix} {first_name} {last_name}"

def add_employee(prefix):
    empid = int(input("Enter the employee id: "))
    first_name = input("Enter the first name: ")
    last_name = input("Enter the last name: ")
    designation = input("Enter the employee designation: ")
    joining_date = input("Enter the employee Joining date: ")
    salary = float(input("Enter the salary: "))

    full_name = generate_full_name(prefix, first_name, last_name)

    employee = {}
    employee["id"] = empid
    employee["name"] = full_name
    employee["job_role"] = designation
    employee["join_date"] = joining_date
    employee["sal"] = salary
    
    employees[empid] = employee

def remove_employee(emp_id):
    if emp_id in employees:
        del employees[emp_id]
        print(f"Employee with ID {emp_id} removed successfully.")
    else:
        print(f"Employee with ID {emp_id} not found.")

employees = {}

# Add initial 5 employees
for i in range(5):
    prefix = validate_prefix()
    add_employee(prefix)

# Add one more employee
prefix = validate_prefix()
add_employee(prefix)

print("Employee Details:")
# Print employee details
for empid, emp_details in employees.items():
    print(f"Employee ID: {empid}")
    for key, value in emp_details.items():
        print(f"{key.capitalize()}: {value}")
    print()

# Remove an employee
remove_id = int(input("Enter the ID of the employee you want to remove: "))
remove_employee(remove_id)

print("Updated Employee Details:")
# Print updated employee details
for empid, emp_details in employees.items():
    print(f"Employee ID: {empid}")
    for key, value in emp_details.items():
        print(f"{key.capitalize()}: {value}")
    print()
```
