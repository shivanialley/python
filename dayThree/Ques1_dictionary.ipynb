```python
#2. check senior most employee from the directory#maxof()

import datetime
def add_employee():
    empid = int(input("Enter the employee id: "))
    emp_name = input("Enter the employee name: ")
    designation = input("Enter the employee designation: ")
    # Handling the input of Joining date format
    while True:
        try:
            joining_date = input("Enter the employee Joining date (YYYY-MM-DD): ")
            datetime.datetime.strptime(joining_date, "%Y-%m-%d")
            break
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")
    salary = float(input("Enter the salary: "))

    employee = {}
    employee["id"] = empid
    employee["name"] = emp_name
    employee["job_role"] = designation
    employee["join_date"] = joining_date
    employee["sal"] = salary
    
    employees[empid] = employee

employees = {}

num=5

for i in range(num):
    add_employee()
    
print("Employee Details:")
for empid, emp_details in employees.items():
    print(f"Employee ID: {empid}")
    for key, value in emp_details.items():
        print(f"{key.capitalize()}: {value}")
    print()

# Find the senior most employee based on joining date
def get_joining_date(emp_id):
    return datetime.datetime.strptime(employees[emp_id]["join_date"], "%Y-%m-%d")

senior_employee_id = max(employees, key=get_joining_date)
senior_employee_details = employees[senior_employee_id]

print("\nSenior Most Employee:")
print(f"Employee ID: {senior_employee_id}")
for key, value in senior_employee_details.items():
    print(f"{key.capitalize()}: {value}")
```
