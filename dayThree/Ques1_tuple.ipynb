```python
#2. check senior most employee from the directory#maxof()
import datetime

def add_employee():
    empid = int(input("Enter the employee id: "))
    emp_name = input("Enter the employee name: ")
    designation = input("Enter the employee designation: ")
    # Handling the input of Joining date format
    while True:
        try:
            joining_date = input("Enter the employee Joining date (YYYY-MM-DD): ")
            datetime.datetime.strptime(joining_date, "%Y-%m-%d")
            break
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")
    salary = float(input("Enter the salary: "))

    employee_tuple = (empid, emp_name, designation, joining_date, salary)
    employees.append(employee_tuple)

employees = []

num = 5

for i in range(num):
    add_employee()

print("Employee Details:")
for emp in employees:
    print(f"Employee ID: {emp[0]}")
    print(f"Name: {emp[1]}")
    print(f"Designation: {emp[2]}")
    print(f"Joining Date: {emp[3]}")
    print(f"Salary: {emp[4]}")
    print()

# Find the senior most employee based on joining date
def get_joining_date(emp):
    return datetime.datetime.strptime(emp[3], "%Y-%m-%d")

senior_employee = max(employees, key=get_joining_date)

print("\nSenior Most Employee:")
print(f"Employee ID: {senior_employee[0]}")
print(f"Name: {senior_employee[1]}")
print(f"Designation: {senior_employee[2]}")
print(f"Joining Date: {senior_employee[3]}")
print(f"Salary: {senior_employee[4]}")
```
