```python
class Employee:
    def __init__(self, id, name, job_role, joining_date):
        self.id = id
        self.name = name
        self.job_role = job_role
        self.joining_date = joining_date
    def display_details(self):
        print("ID:", self.id)
        print("Name:", self.name)
        print("Job Role:", self.job_role)
        print("Joining Date:", self.joining_date)
# 1. Create dictionary, list, and tuple to store employee details
employee_dict = {}
employee_list = []
employee_tuple = ()
# Function to add employee to dictionary
def add_employee_to_dict(employee):
    employee_dict[employee.id] = employee
# Function to add employee to list
def add_employee_to_list(employee):
    employee_list.append(employee)
# Function to add employee to tuple
def add_employee_to_tuple(employee):
    global employee_tuple
    employee_tuple += (employee,)
# 2. Add 5 employees to each data structure
for i in range(1, 6):
    employee = Employee(i, f"Employee{i}", "Role", f"2022-01-0{i}")
    add_employee_to_dict(employee)
    add_employee_to_list(employee)
    add_employee_to_tuple(employee)
# 3. Check senior most employee
senior_employee_dict = employee_dict[max(employee_dict.keys(),key=int)]
senior_employee_list = max(employee_list, key=lambda emp: emp.id)
senior_employee_tuple = max(employee_tuple, key=lambda emp: emp.id)
print("Senior most employee (Dictionary):")
senior_employee_dict.display_details()
print()
print("Senior most employee (List):")
senior_employee_list.display_details()
print()
print("Senior most employee (Tuple):")
senior_employee_tuple.display_details()
print()
# 4. Sort employees by name
sorted_employee_dict = sorted(employee_dict.values(), key=lambda emp:emp.name)
sorted_employee_list = sorted(employee_list, key=lambda emp: emp.name)
sorted_employee_tuple = sorted(employee_tuple, key=lambda emp:emp.name)
print("Employees sorted by name (Dictionary):")
for employee in sorted_employee_dict:
    employee.display_details()
    print()
print("Employees sorted by name (List):")
for employee in sorted_employee_list:
    employee.display_details()
    print()
print("Employees sorted by name (Tuple):")
for employee in sorted_employee_tuple:
    employee.display_details()
    print()
# 5. Remove an employee from the record
eployee_to_remove = employee_list[0]
employee_dict.pop(employee_to_remove.id)
employee_list.remove(employee_to_remove)
employee_tuple = tuple(emp for emp in employee_tuple if emp !=employee_to_remove)
print("Employee removed successfully.")
 6. Function to display prefix, first name, and last name of the employee
def display_employee_details(employee):
    name_parts = employee.name.split()
    print(name_parts)
    prefix = name_parts[0]
   # fname = name_parts[1]
    lname = ' '.join(name_parts[2:])
    print("Prefix:", prefix)
  #  print("First Name:", fname)
    print("Last Name:", lname)
# 7. Print employees with salary above 80,000 (Assuming salary information is not available in the data)
print("\nEmployees with salary above 80,000:")
for employee in employee_list:
    print(employee.name) # Placeholder for actual salary check

# Test function with an employee
print("\nEmployee details (using function):")
display_employee_details(employee_list[0])
```
