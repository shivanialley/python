```python
 """c. Calculate sum of quantity, average of price, total Sales & top selling product
                     [use the following methods: sum and mean]"""
import pandas as pd

df=pd.read_csv("data.csv")
print("Missing Value data")
print(df)
print("\n")
df["Price"].fillna(90000,inplace=True)
df["Quantity"].fillna(20,inplace=True)
df.loc[4,"Product ID"]=1005
df.loc[16,"Product ID"]=1017
print("After handling data")
print(df)
print("Sum of Quantity: ",df["Quantity"].sum())
print("Average of price: ",df["Price"].mean())
df["Sales"]=df.Quantity*df.Price
print(df)
print("Total Sales: ",df["Sales"].sum())
print("Top Selling Product: ",df.loc[df["Sales"].idxmax(),"Product Name"])
```
